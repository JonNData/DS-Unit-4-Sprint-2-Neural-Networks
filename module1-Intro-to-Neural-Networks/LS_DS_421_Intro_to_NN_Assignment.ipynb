### Input Layer:

    Also known as a node, shows the activation which is the sigmoidal(or other function) sum of weighted inputs

    When initialized it can be random. Assign importance to neuron from input to output

    Simplest NN architecture. 
### Loss function
    MSE
    Measure of error, trying to minimize. Early stopping will compare the change in this over epochs

    one pass over your inputs, used to backpropagate

## Write your own perceptron code that can correctly classify (99.0% accuracy) a NAND gate. 

| x1 | x2 | y |
|----|----|---|
| 0  | 0  | 1 |
| 1  | 0  | 1 |
| 0  | 1  | 1 |
| 1  | 1  | 0 |

```python
import pandas as pd
import numpy as np
data = { 'x1': [0,1,0,1],
         'x2': [0,0,1,1],
         'y':  [1,1,1,0],
         'b': [1,1,1,1]
       }

df = pd.DataFrame.from_dict(data).astype('int')
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1-sx)
```

```python
# Initialize random weights

weights = 2 * np.random.random((4,1)) - 1
weights.shape
```

```python
# sum of dot product of inputs and weights
inputs = df
outputs = ([1],[1],[1],[0])
weighted_sum = np.dot(inputs, weights)
```

```python
# first output
activated_output = sigmoid(weighted_sum)
activated_output
```

```python
# Find the error
error = data['y'] - activated_output
error
```

```python
# Gradient descent, finding the steepest slope to lower error

adjustments = error * sigmoid_derivative(weighted_sum)
adjustments.shape
```

```python
# Update the weights 
weights = weights + np.dot(inputs.T, adjustments)
weights
```

```python
# Forward Propagation time

for i in range(10000):
  # dot product the inputs and weights to get weighted sum
  weighted_sum = np.dot(inputs, weights)
  
  # activation function, could add bias here if you wanted
  activated_output = sigmoid(weighted_sum)

  # actual - predicted = error
  error = outputs - activated_output

  # Back Prop time
  adjustments = error * sigmoid_derivative(weighted_sum)

  # Update wthe weights with the dot product of inputs.T and adjustments
  weights = weights + np.dot(inputs.T, adjustments)
print("Weights after training")
print(weights)

print("Output after training")
print(activated_output)
```

```python
import pandas as pd
import numpy as np

diabetes = pd.read_csv('https://raw.githubusercontent.com/ryanleeallred/datasets/master/diabetes.csv')
diabetes.head(10)
```

Although neural networks can handle non-normalized data, scaling or normalizing your data will improve your neural network's learning speed. Try to apply the sklearn `MinMaxScaler` or `Normalizer` to your diabetes dataset. 

```python
from sklearn.preprocessing import MinMaxScaler, Normalizer

features = list(diabetes)[:-1]
target = list(diabetes)[-1:]

# Instantiate normalizer
normalizer = Normalizer()

# Normalize the feature data
X = normalizer.fit_transform(diabetes[features])
# Set target vector
y = diabetes[target]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
class Perceptron:
    
    def __init__(self, niter = 10):
        self.niter = niter
    
    def __sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def __sigmoid_derivative(self, x):
        sx = sigmoid(x)
        return sx * (1-sx)

    def fit(self, X, y):
        """Fit training data
        X : Training vectors, X.shape : [#samples, #features]
        y : Target values, y.shape : [#samples]
        """
        # Randomly Initialize Weights
        self.weight = 2 * np.random.random((8,1)) - 1

        for i in range(self.niter):
            # Weighted sum of inputs / weights
            weighted_sum = np.dot(X, self.weight) + 1
            # Activate!
            activated_output = self.__sigmoid(weighted_sum)
            # Cac error
            error = y - activated_output
            adjustments = error * self.__sigmoid_derivative(weighted_sum)
            # Update the Weights
            self.weight = self.weight + np.dot(X.T, adjustments)

        return self
    
    
    def predict(self, X):
        """Return class label after unit step"""
        weighted_sum = np.dot(X, self.weight) + 1
        activated_output = self.__sigmoid(weighted_sum)
        return np.round(activated_output)
```

```python
model = Perceptron()
model.fit(X, y)
model.predict(X)
#I just for the life of me can't understand shapes and dimensions here...
```
